```python
# import necessary modules
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from classy import Class
from Helper_Functions import Get_LCDM_Pks, Get_DMEFF_Pks, Get_WDM_Pks
```

```python
# create fiducial instance of Class
h=0.6727; omega_b=0.02222; omega_cdm=0.1199; tau_reio=0.078; A_s=2.1985e-9; n_s=0.9652; P_k_max=100
k_vec, LCDM_Pk_vec = Get_LCDM_Pks(h, omega_b, omega_cdm, tau_reio, A_s, n_s, P_k_max)
```

```python
# create interacting instance of Class
omega_dmeff=0.1199; sigma_dmeff=1e-27; m_dmeff=1e-3; npow_dmeff=0.0; dmeff_target='hydrogen'; k_per_decade=40
k_vec, DMEFF_Pk_vec = Get_DMEFF_Pks(h, omega_b, omega_cdm, tau_reio, A_s, n_s, P_k_max, omega_dmeff, \
                                    sigma_dmeff, m_dmeff, npow_dmeff, dmeff_target, k_per_decade)
```

```python
# generate WDM Data
m_dmeff = 0.3 #KeV
k_vec, WDM_Pk_vec = Get_WDM_Pks(m_dmeff, omega_dmeff, k_vec, h)
```

```python
# Plot Transfer Function
plt.plot(k_vec,DMEFF_Pk_vec/LCDM_Pk_vec,'c-',label=r'$\sigma_0 = 10^{-27} cm^{2}$')
plt.plot(k_vec, WDM_Pk_vec,'b:',label=r'$m_{WDM}=%1.1f$ $\rm{KeV}$' %m_dmeff)
# Plot specifications
plt.xscale("log")
plt.xlim(1e-1,k_vec[-1])
plt.ylabel(r'$P_{collisional}/P_{CDM}$',fontsize=12)
plt.xlabel(r'$k\ [h/Mpc]$',fontsize=12)
plt.legend(loc='right',bbox_to_anchor=(1.6, 0.5),fontsize=13)
```
